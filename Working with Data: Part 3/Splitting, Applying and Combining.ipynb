```python
import pandas as pd
from pandas import Series,DataFrame

import numpy as np
```

```python
dframe = pd.read_csv('winequality-red.csv',sep=';')
```

```python
dframe.head()
```

```python
#create a func to rank each wine based on alcohol
def ranker(df):
    df['alcohol_cont_rank']= np.arange(len(df))+1
    return df
```

```python
dframe.sort('alcohol',ascending=False,inplace=True)
```

```python
dframe = dframe.groupby('quality').apply(ranker)

dframe.head()
```

```python
#we can call values whose rank is 1

num_of_qual = dframe['quality'].value_counts()

#Show
num_of_qual
```
